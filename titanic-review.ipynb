---
#### Fill the missing values in the 'age' variable with the median of the age variable.
---

```python
df["age"] = df["age"].fillna(df["age"].median())
df.head()
```

---
#### Check the type of the embarked variable. Change its type to category. Check its type again.
---

```python
df["embarked"].dtype
```

```python
df["embarked"] = df["embarked"].astype("category")
df["embarked"].dtype
```

---
#### Is there a relationship between survival rates and passengers' ages in the Titanic dataset?
---

```python
df["age_cat"] = pd.cut(df["age"], bins=[1, 18, 30, 50, 100], labels=["0-18", "19-30", "31-50", "51+"])
df.head()
a = df.groupby("age_cat", observed=True).agg({"survived" : "sum"})
print(a)
```
